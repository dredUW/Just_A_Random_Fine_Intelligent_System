```python
#imports
import pandas as pd
import numpy as np
import hvplot.pandas
from pathlib import Path
```

```python
# Set the file path
filepath = Path("lstm_daily_return.csv")

# Read the CSV located at the file path into a Pandas DataFrame
stocks = pd.read_csv(filepath)
#set the index as the Date column
stocks.set_index("Date", drop=True, inplace=True)
# Print the DataFrame
stocks.head()
```

```python
# As a test visualize actual exit point relative to actual return
# Using previous daily returns to predict future returns
actual_exit = stocks[stocks['Actual Entry/Exit'] == -1.0]['Actual Return'].hvplot.scatter(
    color='red',
    marker='v',
    size=200,
    legend=False,
    ylabel='Actual Return',
    width=1000,
    height=400
)

# Visualize actual entry point relative to actual return
actual_entry = stocks[stocks['Actual Entry/Exit'] == 1.0]['Actual Return'].hvplot.scatter(
    color='green',
    marker='^',
    size=200,
    legend=False,
    ylabel='Actual Return',
    width=1000,
    height=400
)

# Visualize actual daily return
actual_daily_return = stocks[['Actual Return']].hvplot(
    line_color='lightgray',
    ylabel='Actual Return',
    width=1000,
    height=400
)

# Overlay plots
actual_entry_exit_plot = actual_daily_return * actual_entry * actual_exit
actual_entry_exit_plot.opts(xaxis=None)
```

```python
# Visualize predicted exit point relative to actual return
# Using previous daily returns to predict future returns
predicted_exit = stocks[stocks['Predicted Entry/Exit'] == -1.0]['Actual Return'].hvplot.scatter(
    color='red',
    marker='v',
    size=200,
    legend=False,
    ylabel='Actual Return',
    width=1000,
    height=400
)

# Visualize predicted entry point relative to actual return
predicted_entry = stocks[stocks['Predicted Entry/Exit'] == 1.0]['Actual Return'].hvplot.scatter(
    color='green',
    marker='^',
    size=200,
    legend=False,
    ylabel='Actual Retrun',
    width=1000,
    height=400
)

# Visualize actual daily return
actual_daily_return = stocks[['Actual Return']].hvplot(
    line_color='lightgray',
    ylabel='Actual Return',
    width=1000,
    height=400
)


# Overlay plots
predicted_entry_exit_plot = actual_daily_return * predicted_entry * predicted_exit
predicted_entry_exit_plot.opts(xaxis=None)
```
